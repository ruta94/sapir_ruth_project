# Scrape the Data

We want to scrape the data from this link:
https://www.carmax.com/cars/sedans. 

The basic command with the request package (like seen in our tutorials) does not work but we were able to scrape the data with selenium and a circumvent:
* Click Right -> Inspect -> Network
* then, by clicking on the bottom button "SEE MORE MATCHES" the network tab updates and a url appears. 
* when opening the url in a new tab, we can see the data in a json format.
* this is the url that we will scrape with selenium.

<img src="inspect_website_1.png" />

```python
import pandas as pd
import numpy as np
from selenium import webdriver
import json
from tqdm import tqdm
```

When using the regular package **requests** with the **get** command, we actually got an error that we don't have permission. 

When inspecting (click right) the webpage we discovered that the data is fetchable through their API.

```python
import requests
response = requests.get("https://www.carmax.com/cars/sedans")
response
```

```python
# In the url here below we can change the zip code if we want to scrape from another location.

url = 'view-source:https://www.carmax.com/cars/api/search/run?uri=%2Fcars%2Fall&skip=start&take=100&zipCode=14228&radius=90&sort=20&visitorID=85e86270-42c2-4c91-ab09-d8c865ea5253'
# try opening that link without the view-source part, it will still have some html

# we preliminarily downloaded the chrome driver (we work with chrome)

driver = webdriver.Chrome()
df_main = pd.DataFrame()

# loop that collects the data, you want to put the number of results (rounded to the higher 100th digit)
# like if there are 29030 results, change (0, 31800, 100) to (0, 29100, 100) -- according to the number of results displayed in the website's bottom part.
    
for i in tqdm(range(0, 31800, 100)):
    driver.get(url.replace('start', str(i)))
    # this time with chrome we need to put 'line-content', when clicking right on the html page where the json appears we saw that the line-content is what we needed.
    # It would have been different with another browser.
    data = json.loads(driver.find_element_by_class_name('line-content').text)
    df= pd.DataFrame(data['items'])
    df_main = df_main.append(df)

# and it will save the data in the csv as named below

df_main.reset_index(drop=True).to_csv('NYC-Buffalo.csv')
```

The above procedure was done for two locations, NYC and Kansas City.

### Relation between 'lastMadeSaleableDate' and 'comingSoonDate'

Like shown here below, there is no single field with both null, and as the field names are tranparent, we understand that both field actually include the same kind of information:
* **lastMadeSaleableDate** is actually the date when the car was uploaded to the website for sale. 
* **comingSoonDate** is the date when the vehicle will be available for sale.


```python
data[['comingSoonDate','lastMadeSaleableDate']].head(10)
```

```python
data[(~(data['comingSoonDate'].isnull()) & ~(data['lastMadeSaleableDate'].isnull()))][['comingSoonDate','lastMadeSaleableDate']]
```

```python
(data['comingSoonDate'].isnull() & data['lastMadeSaleableDate']).isnull().sum()
```

We can combine those two columns in a single column counting the number of days between the specified dates and the current date:
* positive numbers will mean that the car was already released for sale.
* negative numbers will mean that the car is yet not available for sale.

```python
from datetime import datetime

data['DaysToSaleable']=np.where(data['lastMadeSaleableDate'].isnull(),data['comingSoonDate'],data['lastMadeSaleableDate'])


def convert_to_timestamp(x):
    split_string = x.split(".")[0]
    x_converted=datetime.strptime(split_string.replace('T',' ').replace('Z',''), "%Y-%m-%d %H:%M:%S")
    return x_converted

data['DaysToSaleable']=data['DaysToSaleable'].map(convert_to_timestamp)

data['DaysToSaleable']=[(datetime(2021,5,28)-x).days for x in data['DaysToSaleable']]
```

We create the function **plot_variable_against_baseprice** which knows how to plot each feature against the predicted feature **basePrice**:
* when the feature is categorical we use the package **seaborn** in order to plot boxplot.
* when the feature is numerical we use **seaborn** to plot a scatterplot.

```python
def plot_variable_against_baseprice(field):
    if (data.dtypes[data.dtypes.index==field]=='object').bool():
        sns.boxplot(x=field, y="basePrice",
            data=data)
        plt.xticks(rotation=90)
        plt.title(f''' Variable '{field}' Against basePrice''')
        sns.set(rc = {'figure.figsize':(15,8)})
    else:
        sns.scatterplot(data=data, x=field, y="basePrice")
        plt.title(f''' Variable '{field}' Against basePrice''')
        sns.set(rc = {'figure.figsize':(15,8)})
```

## Dummify The Categorical Variables

Here below we convert categorica data into dummy variables.

```python
X_cat = data[data.columns[data.dtypes=='object']]


X_dummies = pd.get_dummies(data=X_cat, drop_first=True)
X_dummies.head()
```

```python
X = pd.concat([data[data.columns[data.dtypes!='object']], X_dummies], axis=1)
X=X.drop('basePrice',axis=1)

Y=data['basePrice']
```

```python
(X.isnull().sum(axis = 0)).index[(X.isnull().sum(axis = 0)) >0]
```

## Run the Multiple Linear Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=101)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

The model of multiple linear regression is the following:

$$Y=\alpha_0+\alpha_1 \times X_1+\alpha_2 \times X_2+...+\alpha_n \times X_n$$

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)

# print the intercept
print(model.intercept_)

coeff_parameter = pd.DataFrame(model.coef_,X.columns,columns=['Coefficient'])
coeff_parameter
```

The above are actually the estimated $\alpha_0,\alpha_1 \, \alpha_2 ...\alpha_n $

We see that, as a consequence of the dummification of categorical variables, we get a total number of variables which is more than 800. 

We are aware that it's a lot! and that the prediction may be overfitted (bad impact of too many variables), but it's a start!

```python
predictions = model.predict(X_test)
```

## Check the performances of the model

The package that evaluates the performances of the multiple linear regression ("multiple" because there is more than a single variable).
is different from the packager that runs the regression.

```python
import statsmodels.api as sm
X_train_Sm= sm.add_constant(X_train)
ls=sm.OLS(y_train,X_train_Sm).fit()
print(ls.summary())
```

What should we look at here above?

* $P> \lvert x \lvert$ this is the **p_value** of the test of nullity of each coefficient. If $\alpha_1$ has a p_value lower than 5% it means that he is significantly non-null and hence, that $X_1$ significantly impacts the **basePrice**.

* $R^2$ and $R^{2}_{adj}$: they tell you what's the level of fitting of the model. Fitting of 100% means that you predict the **basePrice** without any error. Fitting of 0% means that you don't predict anything well. $R^{2}_{adj}$ is like $R^2$ where the total number of variables is taken into account. (the higher the number of variables is, the higher the $R^2$ is).
